```python
import pandas as pd
import numpy as np
data =pd.read_csv("/content/NewspaperData.csv")
```

```python

df=pd.DataFrame(data)
```

```python
import statsmodels.formula.api as smf  # Not statsmodels.api
import seaborn as sns
sns.distplot(df['daily'])
sns.distplot(df['sunday'])
```

```python
model=smf.ols('sunday~daily',data=df).fit()
```

```python
sns.regplot(x='daily',y='sunday',data=df)
```

```python
print(model.tvalues, '\n' ,model.pvalues)

```

```python
(model.rsquared,model.rsquared_adj)
```

```python
new_data=pd.Series([222,300])
```

```python
data_pred=pd.DataFrame(new_data,columns=['daily'])
```

```python
model.predict(data_pred)
```
