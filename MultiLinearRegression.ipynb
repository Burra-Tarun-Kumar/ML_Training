```python
def get_standardized_values(vals):
    return (vals-vals.mean())/vals.std()
```

```python
rsq_hp = smf.ols('HP~WT+VOL+SP',data=df).fit().rsquared
vif_hp=1/(1-rsq_hp)
print("rsq_hp", rsq_hp)
print("vif_hp", vif_hp)

rsq_vol = smf.ols('VOL~WT+SP',data=df).fit().rsquared
vif_vol=1/(1-rsq_vol)
print("rsq_vol", rsq_vol)
print("vif_vol", vif_vol)

rsq_wt = smf.ols('WT~SP',data=df).fit().rsquared
vif_wt=1/(1-rsq_wt)
print("rsq_wt", rsq_wt)
print("vif_wt", vif_wt)

rsq_sp = smf.ols('SP~VOL',data=df).fit().rsquared
vif_sp=1/(1-rsq_sp)
print("rsq_sp", rsq_sp)
print("vif_sp", vif_sp)
```
